## 1. What is a Variable?
In Python, a variable is like a **container** or a **label** for a specific piece of information.

When you write code, you need a way to save data (like a score in a game, a user's name, or the price of an item) so you can use it later. Variables allow you to store this data in your computer's memory and give it a name.

* **The Name:** How you refer to the data (e.g., `user_age`).
* **The Value:** The actual data stored (e.g., `25`).

# ðŸ§® Chapter 1.2: Basic Operators

Operators are the symbols we use to perform operations on variables and values. If variables are the **nouns** (the things), operators are the **verbs** (the actions).



## 1. Arithmetic Operators (The Math)
These are used to perform common mathematical operations.

| Operator | Name | Description | Example |
| :--- | :--- | :--- | :--- |
| `+` | Addition | Adds two values | `5 + 2` (7) |
| `-` | Subtraction | Subtracts one value from another | `5 - 2` (3) |
| `*` | Multiplication | Multiplies two values | `5 * 2` (10) |
| `/` | Division | Divides and returns a decimal (float) | `5 / 2` (2.5) |
| `//` | Floor Division | Divides and rounds down to the nearest whole number | `5 // 2` (2) |
| `%` | Modulo | Returns the **remainder** after division | `5 % 2` (1) |
| `**` | Exponentiation | Power (5 to the power of 2) | `5 ** 2` (25) |



---

## 2. The Analogy: The Remainder (`%`)
Think of the **Modulo (`%`)** operator like sharing cookies:
* You have **5 cookies**.
* You have **2 friends**.
* Each friend gets **2 whole cookies** (that is the Floor Division: `5 // 2 = 2`).
* There is **1 cookie left over** that you can't split evenly.
* The **1** is the result of `5 % 2`.

---

## 3. Comparison Operators (The Questions)
Comparison operators are used to compare two values. They **always** return a Boolean: `True` or `False`. Think of them as asking a question.

| Operator | Name | Question | Example |
| :--- | :--- | :--- | :--- |
| `==` | Equal to | Are these exactly the same? | `5 == 5` (True) |
| `!=` | Not equal to | Are these different? | `5 != 3` (True) |
| `>` | Greater than | Is the left side bigger? | `5 > 10` (False) |
| `<` | Less than | Is the left side smaller? | `5 < 10` (True) |
| `>=` | Greater than or equal | Is it bigger or the same? | `5 >= 5` (True) |
| `<=` | Less than or equal | Is it smaller or the same? | `4 <= 2` (False) |



> **âš ï¸ Warning:** Do not confuse `=` with `==`.
> * `=` is **Assignment** (Putting a value in a box).
> * `==` is **Comparison** (Checking if two boxes are the same).

---

2. **E**xponents `**`
3. **M**ultiplication and **D**ivision `* / // %`
4. **A**ddition and **S**ubtraction `+ -`
